```python
#그래프 라이브러리
import matplotlib.pyplot as plt

#한글폰트 적용
plt.rc('font', family= 'Malgun Gothic')

sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']

#제목
plt.title('월별 판매 실적')
plt.plot(month, sales, color='blue', label = "단위:개수") #범례 "단위:개수"
plt.legend(loc = 'lower right') #위치를 정해준다. 아래 오른쪽
#x,y축 제목 월 매출
plt.xlabel('월')
plt.ylabel('매출')

plt.show()
```

```python
#그래프 라이브러리
import matplotlib.pyplot as plt

#한글폰트 적용
plt.rc('font', family= 'Malgun Gothic')

sales = [1,5,7,3,7]
month = [3,4,5,6,7]
# 막대 그래프 plt.bar 가로막대 barh
plt.barh(month, sales)
plt.show()
```

```python
#그래프 라이브러리
import matplotlib.pyplot as plt

#한글폰트 적용
plt.rc('font', family= 'Malgun Gothic')

b_type = [25, 19, 37, 11]
b_name = ['A형','B형','O형','AB형']
#원형 그래프 plt.pie
plt.pie(b_type, labels = b_name,autopct ='%1.1f%%') # 범례추가
plt.show()
```
