```python
print('a b   .')
a = int(input(' a .: '))
b = int(input(' b .: '))

#  
if a > b :
    a, b = b, a
    
sum = 0
for i in range(a , b+1):
    if i < b:
        print(f'{i} + ',end="")
    else:
        print(f'{i} = ',end="")
    sum = sum + i
    
print(sum)
```

```python
print('a b   .')
a = int(input(' a .: '))
b = int(input(' b .: '))

#  
if a > b :
    a, b = b, a
    
sum = 0
for i in range(a, b):
    print(f'{i} + ',end="")
    sum = sum + i
    
print(f'{b} = ', end="")
sum = sum + b

print(sum)
```

```python
print('+ -  .')

n = int(input('  ?: '))

for i in range(n):
    if i % 2:
        print('-',end="")
    else:
        print('+',end="")
print()
```
