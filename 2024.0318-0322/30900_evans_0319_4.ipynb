```python
print('1 n   .')
n = int(input('n .: '))

#  
sum = 0
i = 1

while i <= n: # i n    
    sum += i
    i += 1
    
print(f'1 {n}   {sum}.')
```

```python
# for  .

print('1 n   .')
n = int(input('n .: '))

#  
sum = 0
# for  
for i in range(1, n+1):
    sum += i
    
print(f'1 {n}   {sum}.')
```

```python
print('a b   .')
a = int(input('a .: '))
b = int(input('b .: '))

if a > b :
    a, b = b, a #a,b  

#  
sum = 0

# for  
for i in range(a, b+1):
    sum += i
    
print(f'{a} {b}   {sum}.')
```

```python
# a b       
print('a b   .')
a = int(input('a .: '))
b = int(input('b .: '))

if a > b :
    a, b = b, a #a,b  

#  
sum = 0

for i in range(a, b+1):
    if i < b: # ib     
        print(f'{i} + ',end='')
    else:
        print(f'{i} = ',end='')
    sum += i
    
print(sum)
```
