### 데이터 입력해 데이터 프레임 만들기


```python
# pandas 패키지를 로드하기
import pandas as pd
```

```python
# data frame 만들기
# data frame 변수 = df

df = pd.DataFrame({'name' :['김지훈', '이유진','박동현','김민지'],
                                   'english' :[90,80,60,70],
                                   'math' : [50,60,100,20]})
df
```

```python
# 특정 변수의 값을 추출하기
# 영어 점수만 출력
df['english']
```

```python
# 수학 점수만 출력
df['math']
```

```python
#변수의 값으로 합계구하기
#sum
sum(df['english'])
```

```python
# 수학합계
sum(df['math'])

# 과목별 평균

print(sum(df['english']) / 4)
print(sum(df['math']) /4)
```

### 엑셀 파일 불러오기
#### 1. 엑셀 파일 살펴보기
#### 2. 워킹 디렉터리에 엑셀 파일 삽입하기

```python
# 외부파일 불러오는 메소드 read_excel(''디렉토리/파일명)
# 변수에 할당
df_exam = pd.read_excel('excel_exam.xlsx')
df_exam
```

```python
# 문제 영어 점수 합계를 구하시오
# 문제 영어 점수 평균를 구하시오
print('영여점수 합계', sum(df_exam['english']))
print('영여점수 평균', sum(df_exam['english'])/len(df_exam))

print('수학점수 합계', sum(df_exam['math']))
print('수학점수 평균', sum(df_exam['math'])/len(df_exam))

print('과학점수 합계', sum(df_exam['science']))
print('과학점수 평균', sum(df_exam['science'])/len(df_exam))
```
