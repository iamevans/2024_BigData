```python
## 꺽은선 그래프 그리기
## 라이브러리에 별명을 지정한다. import 라이브러리 ~ as 별명
import matplotlib.pyplot as plt

# 꺽은선 그래프 명령어 plt.plot
plt.plot([1,5,7,3,7])
plt.show()
```

```python
# 라이브러리 불러오기 및 별명 지정
import matplotlib.pyplot as plt

# 그래프용 데이터
sales = [1,5,7,3,7]
month = [3,4,5,6,7]

# 그래프 그리기
plt.plot(month, sales)
plt.show()
```

```python
# 라이브러리 불러오기 및 별명 지정
import matplotlib.pyplot as plt
# 한글 폰트
plt.rc('font', family="Malgun Gothic")
#  그래프 해상도 설정
plt.figure(dpi=100)

# 그래프용 데이터
sales = [1,5,7,3,7]
month = ["mar","apr","may","jun","jul"]

# 한글 폰트
plt.rc('font', family="Malgun Gothic")
# 그래프 제목 넣기 plt.title
plt.title('월별 판매 실적')

# 그래프 그리기
plt.plot(month, sales, color='red') # 그래프 색 바꾸기 color="색이름"
plt.show()
```
