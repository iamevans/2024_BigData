```python
# 1 n   

print('1 n   .')

# n 0  .
while True:
    n = int(input('n .: '))
    if n > 0:
        break
        
#           
sum = 0
i = 1

for i in range(1, n+1):
    sum += i
    i += 1
    
print(f'1 {n}   {sum}.')
```

```python
# ,     area    


area = int(input('  .: '))

for i in range(1, area+1):
    if i * i > area : break
    if area % i :continue
    print(f'{i} x {area // i}')
```
